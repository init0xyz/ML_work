```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("ex1data1.txt",delimiter=',')
X = data[:,0]
X = X.reshape((X.size,1))
y = data[:,1]
y = y.reshape((y.size,1))
plt.plot(X,y,'rx')
plt.ylabel("Profit in $10,000s")
plt.xlabel("Population of City in 10,000s")
```

```python
import ex1
from imp import reload
reload(ex1)
theta = np.zeros((2,1))
iters_num = 1500
alpha = 0.01
theta = ex1.gradientDescent(X,y,theta,alpha,iters_num)
print("Theta computed from gradient descent:\n%f,\n%f" % (theta[0][0],theta[1][0]))
predict1 = np.dot(np.array([1,3.5]).reshape((1,2)),theta)[0][0]
predict1 *= 10000
print("For population = 35,000, we predict a profit of %f \n" % predict1)
m = X.size
now = np.ones(m)
XX = np.append(now,X)
XX = XX.reshape((m,2),order = 'F')

plt.plot(X,np.dot(XX,theta),'-',label = "Linear Regression")
plt.plot(X,y,'rx',label = "Training Data")
plt.legend()
plt.ylabel("Profit in $10,000s")
plt.xlabel("Population of City in 10,000s")
# plt.legend("Training data","Linear regression")
```
